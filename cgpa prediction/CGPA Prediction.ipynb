```python
#Step 1: import libraries
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
```

```python
#step 2: load dataset
df = pd.read_csv("cgpa.csv")
df
```

```python
#step 3: data preprocessing
print(df.isnull().sum())
```

```python
#step 4: split into x and y axis
X= df.drop(columns=['cgpa'], axis=1)
y= df.cgpa
```

```python
#Step 5: split to training and testing (SKIP) B'coz
#Step 6: Call model and train the model
model = DecisionTreeRegressor()

#train
model.fit(X,y)
```

```python
#prediction
study_hours = int(input("enter your study hours"))
play_hours = int(input("Enter your play hours"))
pred = model.predict([[study_hours, play_hours]])
print(pred.item())
```

```python
#Model evaluation
from sklearn.metrics import r2_score
y_pred = model.predict(X)
y_pred
```

```python
acc = r2_score(y_pred,y)
acc * 100
```
