# Overview
Using pretrained model(stable-diffusion-xl-base-1.0) to generate and modify the image based on text.

```python
#Load the model
from diffusers import DiffusionPipeline
import torch

pipe = DiffusionPipeline.from_pretrained(
    #model
    'stabilityai/stable-diffusion-xl-base-1.0',#pretrained model for text to image generation
    #datatype
    torch_dtype = torch.float16,#fetch image pixel array to float values
    #safetensors
    use_safetensors = True,#to avoid error of float image values
    #variants
    variant = 'fp16'
)

#To check cpu or Gpu
pipe.to("cuda")
```

```python
#Sample text image
prompt = "A portrait of the famous cricketer Virat Kohli in a cricket uniform."
negative_prompt = "Avoid including any other people in the scene."
images = pipe(prompt=prompt, negative_prompt=negative_prompt).images[0]

#Display the image
import matplotlib.pyplot as plt
plt.imshow(images)
plt.axis('off')
plt.show()
```
